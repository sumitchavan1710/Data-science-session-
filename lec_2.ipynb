```python
import pandas as pd
import numpy as np
data={'Name':['Alice','Bob','David','Prachi'],
      'Age':[25,np.nan,30,np.nan],
      'Salary':[50000,60000,None,45000]
      }
df=pd.DataFrame(data)
print("Before cleaning:")
print(df)
a="********************"
print(a)
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].median(),inplace=True)
print("After cleaning:")
print(df)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=[10,12,13,12,11,200,14,13,15,12]
sns.boxplot(data)
plt.show()
Q1,Q3=np.percentile(data,[25,75])
IQR=Q3-Q1
lower,upper=Q1-1.5*IQR,Q3+1.5*IQR
outliers=[x for x in data if x<lower or x>upper]
print("Outliers detected:",outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
df=pd.DataFrame({'AGE':[18,22,30,40],
                 'SALARY':[20000,50000,80000,120000]})
scaler=MinMaxScaler()
df_norm=scaler.fit_transform(df)
scaler_std=StandardScaler()
df_std=scaler_std.fit_transform(df)
print("Original Data:\n",df)
print("\nNormalized:\n",df_norm)
print("\nStandardized:\n",df_std)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
f=pd.DataFrame({
'Name':['Alice','Bob','Charlie'],
'Age':[25,32,47],
'Country':['US','India','US'],
'Spend1':[200,150,300],
'Spend2':[100,200,400]
})
f['Total_Spend']=f['Spend1']+f['Spend2']
f['Age_Group']=pd.cut(f['Age'],bins=[0,30,50],labels=['Young','Middle'])
encoder= LabelEncoder()
f['Country_Code']=encoder.fit_transform(f['Country'])
print(f)
```

```python
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
