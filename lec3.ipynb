```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/tips_dataset_350.csv")  # Sample dataset

# Numerical variable
sns.histplot(df["total_bill"], kde=True)
plt.show()

sns.boxplot(x=df["total_bill"])
plt.show()

# Categorical variable
sns.countplot(x="day", data=df)
plt.show()
```

```python
# Numerical vs Numerical
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()

# Categorical vs Numerical
sns.boxplot(x="smoker", y="tip", data=df)
plt.show()

# Categorical vs Categorical
pd.crosstab(df["sex"], df["smoker"])
sns.heatmap(pd.crosstab(df["sex"], df["smoker"]), annot=True, cmap="YlGnBu")
plt.show()
```

```python
# Pairplot
sns.pairplot(df[["total_bill", "tip", "size"]])
plt.show()

# Heatmap for correlation
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

# Grouped boxplot
sns.boxplot(x="day", y="total_bill", hue="sex", data=df)
plt.show()
```

```python
# Matplotlib basics
plt.plot(df["total_bill"])
plt.title("Line Plot of Total Bills")
plt.show()

# Bar plot
plt.bar(df["day"].value_counts().index, df["day"].value_counts().values)
plt.show()

# Pie chart
df["day"].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
# Descriptive statistics
print(df["total_bill"].describe())
print("Skewness:", df["total_bill"].skew())
print("Kurtosis:", df["total_bill"].kurt())

# Correlation
print(df.corr(numeric_only=True))

# Simple regression
import statsmodels.api as sm

X = df["total_bill"]
y = df["tip"]
X = sm.add_constant(X)  # add intercept
model = sm.OLS(y, X).fit()
print(model.summary())
```
